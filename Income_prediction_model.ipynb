```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

le = LabelEncoder()
le.fit(df['income'])
```

```python
le.classes_
```

```python
df['income']=le.transform(df['income'])
```

```python
le.fit(df['education'])
```

```python
le.classes_
```

```python
df['education']=le.transform(df['education'])
```

```python
workclass=pd.get_dummies(df['workclass'])
df=pd.concat([df,workclass],axis=1)
df.drop(['workclass'],axis=1,inplace=True)

marital_status=pd.get_dummies(df['marital.status'])
df=pd.concat([df,marital_status],axis=1)
df.drop(['marital.status'],axis=1,inplace=True)

relationship=pd.get_dummies(df['relationship'])
df=pd.concat([df,relationship],axis=1)
df.drop(['relationship'],axis=1,inplace=True)

sex=pd.get_dummies(df['sex'])
df=pd.concat([df,sex],axis=1)
df.drop(['sex'],axis=1,inplace=True)

race=pd.get_dummies(df['race'])
df=pd.concat([df,race],axis=1)
df.drop(['race'],axis=1,inplace=True)

occupation=pd.get_dummies(df['occupation'])
df=pd.concat([df,occupation],axis=1)
df.drop(['occupation'],axis=1,inplace=True)

native_country=pd.get_dummies(df['native.country'])
df=pd.concat([df,native_country],axis=1)
df.drop(['native.country'],axis=1,inplace=True)

df
```

```python
X =df.drop(['income'],axis=1)
y =df['income']
```

```python
education=X['education']
X.drop(['education'],axis=1,inplace=True)
X=pd.concat([X,education],axis=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=56)-
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train.iloc[:,:6] = sc.fit_transform(X_train.iloc[:,:6] )
X_test.iloc[:,:6]  = sc.transform(X_test.iloc[:,:6] )
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.metrics import classification_report

target_names = ['<=50K','>50K']
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression(random_state=0)
logistic_regression.fit(X_train, y_train)

y_pred = logistic_regression.predict(X_test)

print("train score = "+ str(logistic_regression.score(X_train, y_train)))
print("test score = "+str(logistic_regression.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn import svm
SVM = svm.SVC()
SVM.fit(X_train, y_train)

y_pred=SVM.predict(X_test)

print("train score = "+ str(SVM.score(X_train, y_train)))
print("test score = "+str(SVM.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn import tree
Decision_tree = tree.DecisionTreeClassifier()
Decision_tree.fit(X_train, y_train)

y_pred=Decision_tree.predict(X_test)

print("train score = "+ str(Decision_tree.score(X_train, y_train)))
print("test score = "+str(Decision_tree.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.naive_bayes import BernoulliNB

naive_bayes= BernoulliNB()

naive_bayes.fit(X_train, y_train)

y_pred=naive_bayes.predict(X_test)

print("train score = "+ str(naive_bayes.score(X_train, y_train)))
print("test score = "+str(naive_bayes.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=2500)

KNN.fit(X_train,y_train)

y_pred=KNN.predict(X_test)

print("train score = "+ str(KNN.score(X_train, y_train)))
print("test score = "+str(KNN.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=100, random_state=10)

random_forest.fit(X_train, y_train)

y_pred=random_forest.predict(X_test)

print("train score = "+ str(random_forest.score(X_train, y_train)))
print("test score = "+str(random_forest.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_boost = AdaBoostClassifier(n_estimators=100, random_state=10)

ada_boost.fit(X_train, y_train)

y_pred=ada_boost.predict(X_test)

print("train score = "+ str(ada_boost.score(X_train, y_train)))
print("test score = "+str(ada_boost.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.ensemble import BaggingClassifier

bagging=BaggingClassifier(estimator=svm.SVC(), n_estimators=10, random_state=0)

bagging.fit(X_train, y_train)

y_pred=bagging.predict(X_test)

print("train score = "+ str(bagging.score(X_train, y_train)))
print("test score = "+str(bagging.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gradient_boosting = GradientBoostingClassifier(n_estimators=100, random_state=10)

gradient_boosting.fit(X_train, y_train)

y_pred=gradient_boosting.predict(X_test)

print("train score = "+ str(gradient_boosting.score(X_train, y_train)))
print("test score = "+str(gradient_boosting.score(X_test, y_test)))

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1]}
gscv=GridSearchCV(estimator=gradient_boosting, param_grid=param_grid, cv=5)

gscv.fit(X_train, y_train)
```

```python
y_pred=gscv.best_estimator_.predict(X_test)
print("Best score: {}".format(gscv.best_score_))
print("Best parameters: {}".format(gscv.best_params_))
print(classification_report(y_test, y_pred, target_names=target_names))
```
